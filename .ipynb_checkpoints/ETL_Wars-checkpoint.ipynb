```python
import pandas as pd
```

```python
war1_path = "interstate-wars/extrastate.csv"
war2_path = "interstate-wars/interstate.csv"
war3_path = "interstate-wars/intrastate.csv"
```

```python
war1_df = pd.read_csv(war1_path)
war1_df
```

```python
war1_a_cleaned = pd.DataFrame({'War ID': war1_df.war_id,
                             'War Name': war1_df.war_name,
                             'Country': war1_df.side1_name,
                             'Start Year': war1_df.start_year1,
                             'End Year': war1_df.end_year1})
war1_a_cleaned
```

```python
war1_b_cleaned = pd.DataFrame({'War ID': war1_df.war_id,
                             'War Name': war1_df.war_name,
                             'Country': war1_df.side2_name,
                             'Start Year': war1_df.start_year1,
                             'End Year': war1_df.end_year1})
war1_b_cleaned
```

```python
frames = [war1_a_cleaned, war1_b_cleaned]
```

```python
war1_df = pd.concat(frames)
war1_df
```

```python
war2_df = pd.read_csv(war2_path)
war2_df
```

```python
wars2_cleaned = pd.DataFrame({'War ID': war2_df.war_id,
                             'War Name': war2_df.war_name,
                             'Country': war2_df.state_name,
                             'Start Year': war2_df.start_year1,
                             'End Year': war2_df.end_year1})
wars2_cleaned
```

```python
war3_df = pd.read_csv(war3_path)
war3_df
```

```python
war3_a_cleaned = pd.DataFrame({'War ID': war3_df.war_id,
                             'War Name': war3_df.war_name,
                             'Country': war3_df.side1_name,
                             'Start Year': war3_df.start_year1,
                             'End Year': war3_df.end_year1})
war3_a_cleaned
```

```python
war3_b_cleaned = pd.DataFrame({'War ID': war3_df.war_id,
                             'War Name': war3_df.war_name,
                             'Country': war3_df.side2_name,
                             'Start Year': war3_df.start_year1,
                             'End Year': war3_df.end_year1})
war3_b_cleaned
```

```python
frames3 = [war3_a_cleaned, war3_b_cleaned]
```

```python
war3_df = pd.concat(frames3)
war3_df
```

```python
wars_list = [war1_df, wars2_cleaned, war3_df]
```

```python
wars_df = pd.concat(wars_list, axis=0, join='outer', ignore_index=False, keys=None,
          levels=None, names=None, verify_integrity=False, copy=True)
wars_df
```

```python
wars_df.dropna()
wars_df
```

```python
wars_df_cleaned = wars_df.loc[wars_df["End Year"] != -7]
wars_df_cleaned
```

```python
wars_df_cleaned_2 = wars_df_cleaned.loc[wars_df_cleaned["Country"] != '-8']
wars_df_cleaned_2
```

```python
wars_df_cleaned_3 = wars_df_cleaned_2.loc[wars_df_cleaned_2['End Year'] !='0']
wars_df_cleaned_3
```

```python
wars_df_cleaned_4 = wars_df_cleaned_3.dropna()
wars_df_cleaned_4
```

```python
wars_df_cleaned_4['End Year'] = wars_df_cleaned_4['End Year'].astype(int)
wars_df_cleaned_4
# wars_df_cleaned_5 = pd.DataFrame({'War ID': wars_df_cleaned_4.War ID,
#                              'War Name': wars_df_cleaned_4.war_name,
#                              'Country': wars_df_cleaned_4.side2_name,
#                              'Start Year': wars_df_cleaned_4.start_year1,
#                              'End Year': int(wars_df_cleaned_4.end_year1)})
# wars_df_cleaned_5
```
