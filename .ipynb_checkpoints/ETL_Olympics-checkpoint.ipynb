```python
#import Dependencies
import pandas as pd
```

```python
#Reading CSV
csv_winter_path = "olympic-games/winter.csv"
csv_summer_path = "olympic-games/summer.csv"
csv_dict_path = "olympic-games/dictionary.csv"
```

```python
#Creating Dataframes with CSV data
winter_df = pd.read_csv(csv_winter_path, encoding="utf-8")
winter_df
```

```python
summer_df = pd.read_csv(csv_summer_path, encoding='utf-8')
summer_df
```

```python
dict_df = pd.read_csv(csv_dict_path, encoding='utf-8')
dict_df
```

```python
# Concatenating Olympic games data into one Dataframe in ascending order)
olympic_list = [summer_df, winter_df]
```

```python
olympic2_df = pd.concat(olympic_list, axis=0, join='outer', ignore_index=False, keys=None,
          levels=None, names=None, verify_integrity=False, copy=True)
olympic2_df
```

```python
# Unifying column name "Code" 
olympic3_df = olympic2_df.rename(columns={'Country':'Code'})
olympic3_df
```

```python
# Adding a new column Country with mapping it to a dictionary dataframe
olympic3_df['Country'] = olympic3_df['Code'].map(dict_df.set_index('Code')['Country'])
olympic3_df
```

```python
# Getting a final dataframe
olympic_final_df = olympic3_df[["Year","Country","Medal","City","Sport","Discipline","Athlete","Gender","Event"]]
olympic_final_df
```

```python
lookup = olympic_final_df.groupby(['Year', 'Country']).size().reset_index()[['Year','Country']].to_dict(orient='records')
```
